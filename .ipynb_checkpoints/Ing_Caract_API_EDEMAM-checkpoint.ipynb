```python
df_total = pd.DataFrame()
Food_name_list = []
for i, element_hits in enumerate (df['hits']):
    Food_name = element_hits['recipe']['label']
    Food_name_list.append(Food_name)
    df_ = pd.DataFrame(element_hits['recipe']['totalNutrients']).T.rename_axis('Nutri_keys')
    df_copy = df_.copy()
    df_rename = df_copy.rename(columns = {'quantity': 'quantity: ' + str(i)})
    df_2 = df_rename.set_index(['label', 'unit'], append = True)
    if i != 0:
        #print("")
        df_total = pd.merge(df_total, df_2, on = ['Nutri_keys', 'label', 'unit'], how = 'inner')
    else:
        df_total = df_2
    #df_ = df_.rename_axis(Food_name, axis = 'rows')
    #df_total = df_total.join(df_2)
    #break
df_total
```
